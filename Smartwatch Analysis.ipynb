```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches

# Assuming your data is loaded into 'df' and has columns 'activity' and 'gender'
df = pd.read_csv('smartwatch.csv')  # Replace 'your_data.csv' with your actual file name

# Group data by activity and gender, then count occurrences
activity_gender_counts = df.groupby(['activity', 'gender'])['gender'].count().reset_index(name='count')

# Create a bar plot with color-coded gender
plt.figure(figsize=(12, 6))
sns.barplot(x='activity', y='count', hue='gender', data=activity_gender_counts,
            palette={0: 'lightcoral', 1: 'skyblue'})

plt.title('Activity Distribution by Gender')
plt.xlabel('Activity')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping

# Manually create legend with colored patches
female_patch = mpatches.Patch(color='lightcoral', label='Female')
male_patch = mpatches.Patch(color='skyblue', label='Male')
plt.legend(handles=[female_patch, male_patch], title='Gender')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the dataset
df = pd.read_csv('smartwatch.csv')

# Group data by gender and calculate the average for steps, calories, and distance
gender_stats = df.groupby('gender')[['steps', 'calories', 'distance']].mean().reset_index()

# Melt the DataFrame to create a suitable format for grouped bar chart
melted_data = pd.melt(gender_stats, id_vars=['gender'], value_vars=['steps', 'calories', 'distance'],
                      var_name='Metric', value_name='Average')

# Create the grouped bar chart
plt.figure(figsize=(10, 6))
# Changed the keys in the palette dictionary to integers 0 and 1
sns.barplot(x='Metric', y='Average', hue='gender', data=melted_data, palette={0: 'lightcoral', 1: 'skyblue'})
plt.title('Steps, Calories, and Distance by Gender')
plt.xlabel('Metric')
plt.ylabel('Average')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

# Manually create legend with colored patches
import matplotlib.patches as mpatches  # Importing here for clarity
female_patch = mpatches.Patch(color='lightcoral', label='Female')
male_patch = mpatches.Patch(color='skyblue', label='Male')
plt.legend(handles=[female_patch, male_patch], title='Gender')

plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('smartwatch.csv')

# Rename the column 'hear_rate' to 'Heart_rate'
if 'hear_rate' in df.columns:
    df = df.rename(columns={'hear_rate': 'Heart_rate'})

# Drop rows with missing Heart_rate values
df = df.dropna(subset=['Heart_rate'])

# Debug: Check unique values in the 'gender' column
print("Unique values in 'gender':", df['gender'].unique())

# Create the KDE plot
plt.figure(figsize=(10, 6))
sns.kdeplot(
    data=df,
    x='Heart_rate',
    hue='gender',
    fill=True,
    common_norm=False,
    palette={0: 'lightcoral', 1: 'skyblue'}  # Use integer keys
)

plt.title('Heart Rate Distribution by Gender')
plt.xlabel('Heart Rate (BPM)')
plt.ylabel('Density')

# Customize legend labels
plt.legend(title='Gender', labels=['Male', 'Female'])

plt.show()
```

# **Heart Rate Distribution by Gender:**
The KDE (Kernel Density Estimate) plot illustrates the distribution of heart rates (in BPM) by gender:

**1. Color Coding:**

Females are represented in light coral, and males in sky blue.

**2. Observations:**

* The female heart rate distribution shows a peak at lower BPM values, indicating that females in the dataset tend to have a lower average heart rate.
* The male heart rate distribution is more spread out and shifted towards higher BPM values, suggesting higher average heart rates or more variation in heart rate across the male participants.
* The densities of heart rates are separated with males showing a broader distribution, while females concentrate more around a lower heart rate.
* Both distributions have skewness, but the female distribution seems more compact compared to the male distribution.

This plot provides insights into the differences in heart rate patterns between males and females, highlighting that gender may influence cardiovascular metrics.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('smartwatch.csv')

# Rename the column 'hear_rate' to 'Heart_rate' if it exists
if 'hear_rate' in df.columns:
    df = df.rename(columns={'hear_rate': 'Heart_rate'})
else:
    print("Column 'hear_rate' not found, assuming 'Heart_rate' already exists.")

# Select the columns for the pair plot
columns_for_pairplot = ['steps', 'distance', 'calories', 'Heart_rate']

# Ensure 'gender' column exists before creating the pair plot
if 'gender' not in df.columns:
    raise KeyError("The 'gender' column is not found in the DataFrame. Please check your data.")

# Create the pair plot
sns.pairplot(df[columns_for_pairplot + ['gender']], hue='gender', palette={0: 'lightcoral', 1: 'skyblue'})  # Include 'gender' in the data for hue
plt.suptitle('Pairwise Relationships between Steps, Distance, Calories, and Heart Rate', y=1.02)
plt.show()
```

**2. Observations:**

* There is a noticeable positive correlation between steps and distance for both genders. As the number of steps increases, so does the distance covered.

**Steps and Calories:**

**Distance and Calories:**
